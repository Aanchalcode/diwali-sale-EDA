**EDA Begins**

**GENDER ANALYSIS**

```python
ax=sns.countplot(data=data,x='Gender')
```

```python
ax=sns.countplot(data=data,x='Gender')
for bars in ax.containers:
    ax.bar_lable(bars)
plt.show()
```

```python


ax = sns.countplot(data=data, x='Gender')
# Add labels to the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')

plt.show()

```

```python
datas=data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
datas
```

```python
datas=data.groupby(['Gender'])['Amount'].sum().sort_values(ascending=False)
datas
```

```python
sns.barplot(data=datas,x='Gender',y='Amount')
```

**AGE GROUP ANALYSIS**

```python
sns.countplot(data=data,x='Age Group',hue='Gender')
plt.xticks(rotation=90)
plt.show()
```

```python
av=sns.countplot(data=data,x='Age Group',hue='Gender')
plt.xticks(rotation=90)

for p in av.patches:
    av.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')

plt.show()
```

as_index=False: This is an optional parameter in groupby() that specifies whether the grouping column(s)should be set as the index of the resulting DataFrame. By setting it to False,you are telling pandas not to set the grouping column as the index, but to keep it as a regular column.

```python
df=data.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False)
df
```

```python
plt.figure(figsize=(10, 6))
df=data.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False)
sns.barplot(data=df, x='Age Group', y='Amount')
plt.show()
```

**STATE ANALYSIS**

```python
plt.figure(figsize=(17,5))
sales_state=data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
sales_state
sns.barplot(data=sales_state,x='State',y='Orders')
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(17,5))
sales_state=data.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_state
sns.barplot(data=sales_state,x='State',y='Amount')
plt.xticks(rotation=30)
plt.show()
```

hence from the above graph we can see that most of the orders and amounts are from UP, MAHARASTRA, KARNATAKA


**MARITAL STATUS**

```python
marital=data.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
av=sns.barplot(data=marital,x='Marital_Status',y='Amount')

for p in av.patches:
    av.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')

plt.show()
```
